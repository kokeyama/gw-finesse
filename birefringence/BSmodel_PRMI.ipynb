```python
import numpy as np                         # Importing numpy
import matplotlib                          # For plotting
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from pykat import finesse                  # Importing the pykat.finesse package
from pykat.commands import *               # Importing all packages in pykat.commands.
from pykat.optics.maps import *            # Importing maps package
from IPython.display import display, HTML  # Allows us to display HTML.
```

```python
base = None
base = finesse.kat()
base.verbose=False
base.load("BSmodel_PRMI.kat")
```

## Here we tune PRM or PRMp separately.

```python
fig = plt.figure(figsize = (16,6))

kat=None
kat=base.deepcopy()
kat.parse("""
xaxis tuning phi lin -180 180 1000
put PRM phi $x1
""")
out = None
out = kat.run()

ax1 = fig.add_subplot(1, 2, 1)
plt.semilogy(out.x, out["CR0"],label="Carrier")
plt.semilogy(out.x, out["SB1"],label="f1")
plt.semilogy(out.x, out["SB2"],label="f2")
plt.semilogy(out.x, out["SB3"],label="f3")
plt.xlabel("PRM tuning [deg]")
plt.ylabel("Amplitude [a.u.]")
plt.legend(loc=0)
plt.title("Power of s-pol in PRC")
plt.grid()

kat=None
kat=base.deepcopy()
kat.parse("""
xaxis tuning phi lin -180 180 1000
put PRMp phi $mx1
""")
out = None
out = kat.run()

ax2 = fig.add_subplot(1, 2, 2)
plt.semilogy(out.x, out["CR0p"],label="Carrier")
plt.semilogy(out.x, out["SB1p"],label="f1")
plt.semilogy(out.x, out["SB2p"],label="f2")
plt.semilogy(out.x, out["SB3p"],label="f3")
plt.xlabel("PRMp tuning [deg]")
plt.ylabel("Amplitude [a.u.]")
plt.legend(loc=0)
plt.title("Power of p-pol in PRC")
plt.grid()

plt.show()
```

## Here we actuate PRM and PRMp at the same time.

```python
fig = plt.figure(figsize = (16,6))

kat=None
kat=base.deepcopy()
kat.parse("""
xaxis tuning phi lin -180 180 1000
put PRM phi $x1
put PRMp phi $mx1
""")
out = None
out = kat.run()

ax1 = fig.add_subplot(1, 2, 1)
plt.semilogy(out.x, out["CR0"],label="Carrier")
plt.semilogy(out.x, out["SB1"],label="f1")
plt.semilogy(out.x, out["SB2"],label="f2")
plt.semilogy(out.x, out["SB3"],label="f3")
plt.xlabel("PRM tuning [deg]")
plt.ylabel("Amplitude [a.u.]")
plt.legend(loc=0)
plt.title("Power of s-pol in PRC")
plt.grid()

kat=None
kat=base.deepcopy()
kat.parse("""
xaxis tuning phi lin -180 180 1000
put PRM phi $x1
put PRMp phi $mx1
""")
out = None
out = kat.run()

ax2 = fig.add_subplot(1, 2, 2)
plt.semilogy(out.x, out["CR0p"],label="Carrier")
plt.semilogy(out.x, out["SB1p"],label="f1")
plt.semilogy(out.x, out["SB2p"],label="f2")
plt.semilogy(out.x, out["SB3p"],label="f3")
plt.xlabel("PRM tuning [deg]")
plt.ylabel("Amplitude [a.u.]")
plt.legend(loc=0)
plt.title("Power of p-pol in PRC")
plt.grid()

plt.show()
```
